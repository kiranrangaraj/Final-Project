```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/Users/kiranrangaraj/Desktop/Final-Project-Used-Car-Price-Predictor/data/vehiclesFinal.csv')
df['year'] = df['year'].astype('int64')
df.drop(['id', 'lat', 'long'],axis = 1,inplace = True)
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df2=df.sample(100)
df2.shape
```

```python
sns.pairplot(df2)
plt.title('Pair Plot')
plt.tight_layout()
plt.savefig('Insight-Pair-Plot',dpi=None)
plt.show()
```

```python
sns.distplot(df.price)
plt.tight_layout()
plt.title("Distribution of Price")
plt.savefig('Insight-Dist-Plot',dpi=None)
plt.show()
```

```python
fig,axs=plt.subplots()
fig.set_size_inches(5,5)
p1=sns.barplot(x='fuel',y='price',data=df,ax=axs)
axs.set_title('Price vs. Fuel Type')
plt.savefig('fig8',dpi=None)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='fuel',y='price',hue='condition',data=df)
plt.title('Price vs. Fuel & Condition')
plt.savefig('fig9',dpi=None)
plt.show()
```

```python
fig,axs=plt.subplots(nrows=2)
fig.set_size_inches(20,10)
sns.barplot(x='year',y='price',data=df,ax=axs[0])
sns.barplot(df.groupby('year').count()['price'].index,df.groupby('year').count()['price'].values,ax=axs[1])
axs[0].set_title('Price vs. Year')
axs[1].set_title('Number of Cars vs. Year')
axs[1].set_ylabel('number of cars')
plt.tight_layout()
plt.savefig('fig1.jpg',dpi=600)
plt.show()
```

```python
fig,axs=plt.subplots(ncols=2)
fig.set_size_inches(12,4)
sns.barplot(x='condition',y='price',data=df,ax=axs[0])
sns.barplot(x='condition',y='price',hue='size',data=df,ax=axs[1])
axs[0].set_title('Price vs. Condition')
axs[1].set_title('Price vs. Condition & Size')
plt.tight_layout()
plt.savefig('fig3.jpg',dpi=600)
plt.show()
```

```python
fig,axs=plt.subplots(ncols=2)
fig.set_size_inches(12,4)
sns.barplot(x='transmission',y='price',data=df,ax=axs[0])
sns.barplot(x='transmission',y='price',hue='size',data=df,ax=axs[1])
axs[0].set_title('Price vs. Transmission')
axs[1].set_title('Price vs. Transmission & Size')
plt.tight_layout()
plt.savefig('fig4.jpg',dpi=600)
plt.show()
```

```python
fig,axs=plt.subplots(nrows=2)
fig.set_size_inches(15,8)
sns.barplot(x='type',y='price',data=df,ax=axs[0])
sns.barplot(x='manufacturer',y='price',data=df,ax=axs[1])
axs[1].set_xticklabels(df.manufacturer.unique(),rotation=91)
axs[0].set_title('Price vs. Type')
axs[1].set_title('Price vs. Manufacturer')
plt.tight_layout()
plt.savefig('fig5.jpg',dpi=600)
plt.show()
```

```python
fig,axs=plt.subplots(ncols=2)
fig.set_size_inches(12,4)
sns.barplot(x='size',y='price',data=df,ax=axs[0])
sns.barplot(x='size',y='price',hue='drive',data=df,ax=axs[1])
axs[0].set_title('Price vs. Size')
axs[1].set_title('Price vs. Size & Drive')
plt.tight_layout()
plt.savefig('fig6.jpg',dpi=600)
plt.show()
```

```python
fig,axs=plt.subplots(ncols=2)
fig.set_size_inches(12,4)
sns.barplot(x='drive',y='price',data=df,ax=axs[0])
sns.barplot(x='drive',y='price',hue='size',data=df,ax=axs[1])
axs[0].set_title('Price vs. Drive')
axs[1].set_title('Price vs. Drive & Size')
plt.tight_layout()
plt.savefig('fig7.jpg',dpi=600)
plt.show()
```

```python
fig,axs=plt.subplots(ncols=2)
fig.set_size_inches(10,4)
sns.barplot(df.groupby('condition').count()['price'].index,df.groupby('condition').count()['price'].values,ax=axs[0])
sns.barplot(df.groupby('transmission').count()['price'].index,df.groupby('transmission').count()['price'].values,ax=axs[1])
axs[0].set_ylabel('number of cars')
axs[0].set_title('Number of Cars vs. Condition')
axs[1].set_ylabel('number of cars')
axs[1].set_title('Number of Cars vs. Transmission')
plt.tight_layout()
plt.savefig('fig10.jpg',dpi=600)
plt.show()
```

```python
fig,axs=plt.subplots(nrows=2)
fig.set_size_inches(16,10)
sns.barplot(df.groupby('year').count()['price'].index,df.groupby('year').count()['price'].values,ax=axs[0])
sns.barplot(df.groupby('manufacturer').count()['price'].index,df.groupby('manufacturer').count()['price'].values,ax=axs[1])
axs[0].set_ylabel('number of cars')
axs[0].set_title('Number of Cars vs. Year')
axs[1].set_ylabel('number of cars')
axs[1].set_title('Number of Cars vs. Manufacturer')
axs[1].set_xticklabels(df.groupby('manufacturer').count()['price'].index,rotation=90)
plt.tight_layout()
plt.savefig('fig12.jpg',dpi=600)
plt.show()
```

```python
fig,axs=plt.subplots(ncols=2)
fig.set_size_inches(10,5)
sns.barplot(df.groupby('fuel').count()['price'].index,df.groupby('fuel').count()['price'].values,ax=axs[0])
sns.barplot(df.groupby('size').count()['price'].index,df.groupby('size').count()['price'].values,ax=axs[1])
axs[0].set_ylabel('number of cars')
axs[0].set_title('Number of Cars vs. Fuel Type')
axs[1].set_ylabel('umber of cars')
axs[1].set_title('Number of Cars vs. Size')
plt.tight_layout()
plt.savefig('fig11.jpg',dpi=600)
plt.show()
```
